```python
sales['Total Sales'] = sales['Quantity'] * sales['Sales'] * (1 - sales['Discount'])
sales.head()
```

```python
import matplotlib.pyplot as plt

# Scatter plot of Sales Quantity vs Total Sales
plt.figure(figsize=(10, 6))
plt.scatter(sales['Quantity'], sales['Total Sales'], c='red', label='Total Sales')
plt.title('Sales Quantity vs Total Sales')
plt.xlabel('Sales Quantity')
plt.ylabel('Total Sales')
plt.legend()
plt.show()

# Bar plot of Total Sales including Shipping Cost
sales['Total Sales Including Shipping'] = sales['Total Sales'] + sales['Shipping Cost']
plt.figure(figsize=(10, 6))
plt.bar(sales.index, sales['Total Sales Including Shipping'], color='blue')
plt.title('Total Sales Including Shipping Cost')
plt.xlabel('Index')
plt.ylabel('Total Sales Including Shipping')
plt.show()
```

```python
import seaborn as sns

# Scatter plot with regression line of Unit Price vs Total Sales
plt.figure(figsize=(10, 6))
sns.regplot(x='Sales', y='Total Sales', data=sales, scatter_kws={'s': 50}, line_kws={'color': 'red'})
plt.title('Unit Price vs Total Sales')
plt.xlabel('Unit Price')
plt.ylabel('Total Sales')
plt.show()

# Box plot of Total Sales grouped by Discount Rate ranges
sales['Discount Rate Category'] = pd.cut(sales['Discount'], bins=[0, 0.1, 0.2, 0.3], labels=['Low', 'Medium', 'High'])
plt.figure(figsize=(10, 6))
sns.boxplot(x='Discount Rate Category', y='Total Sales', data=sales)
plt.title('Total Sales by Discount Rate Category')
plt.xlabel('Discount Rate Category')
plt.ylabel('Total Sales')
plt.show()
```
